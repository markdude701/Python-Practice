```python
#Rock Paper Scissors
import random

true = True
false = False

def checkInput(uInput):
    if uInput == "R":
        return true
    elif uInput == "P":
        return true
    elif uInput == "S":
        return true
    else:
        return false    
    
#Checks if player has won
def compMove():
    print()
    move = random.randint(1,3)
    moveL = ""
    if move == 1:
        moveL = "R"
    elif move == 2:
        moveL = "P"
    elif move == 3:
        moveL = "S"
    print("Computer chose: " + moveL)
    return moveL

def checkWin(player, computer):
    if player == "R" and computer =="R" or player == "S" and computer == "S" or player == "P" and computer == "P":
        return "Draw"
    elif player == "R" and computer == "S" or player == "S" and computer == "P" or player == "P" and computer == "R":
        return "Player Wins"
    elif player == "P" and computer == "S" or player == "R" and computer == "P" or player == "S" and computer == "R":
        return "Computer Wins"

def gameLoop():
    won = false
    
    #GameLoop
    while(won == false):
        print("Welcome to Rock Paper Scissors")
        playerInput = input("Select R / P / S : ")
        correctInput = checkInput(playerInput)
        print("You chose: " + playerInput)
        #print(correctInput)

        computerMove = compMove()
        print(checkWin(playerInput, computerMove))
        
        playA = input("Would you like to play again? (Y/N)")
        if playA == "Y":
            won = false
        else:
            won = true
            print("Thanks for playing! Goodbye!")
            
gameLoop()
```
